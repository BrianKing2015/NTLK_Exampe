```python
import re
import pandas as pd
import os
```

```python
def fileSearch (starts = '', ends = '', rootDir = 'D:\\Gutenberg\\2006\\rawText'):
    temp = []
    for dirName, subdirList, fileList in os.walk(rootDir):

        for fname in fileList:
            if fname.lower().startswith((starts)) & fname.lower().endswith((ends)) :
                temp.append (os.path.join(dirName,fname))
    return temp
```

```python
def findName (bookTarget):
    with open (bookTarget, 'r', errors = "surrogateescape") as f:
        book = f.read()
    book = ''.join([x for x in book if ord(x) < 128])
    lang = re.compile(r"Language: English\n")
    if lang.search(book) != None:
        reg = re.compile(r"\*\*\* START OF THIS PROJECT GUTENBERG EBOOK")
        reg2 = re.compile(r"\s\*\*\*\n")
        regAlt = re.compile(r"\*\*\*START OF THE PROJECT GUTENBERG EBOOK")
        reg2Alt = re.compile(r"\*\*\*\n")
        if reg.search(book) != None:
            targetStart = reg.search(book).span()[1]
            targetEnd = reg2.search(book).span()[0]
            targetName = book[targetStart: targetEnd]
        elif regAlt.search(book) != None:
            targetStart = regAlt.search(book).span()[1]
            targetEnd = reg2Alt.search(book).span()[0]
            targetName = book[targetStart: targetEnd]
        else:
            targetStart = None
            targetEnd = None
            targetName = 'Not Available'
    else:
        targetName = 'Not English'
        
    return targetName, bookTarget

#thing = targetStart.span()[1]
#thing2 = targetEnd.span()[0]
#print (book [targetStart: targetEnd])
#print (book[535:563])

```

```python
thing = fileSearch ('','-8.txt', rootDir = 'D:\\Gutenberg\\2006\\rawText\\')
```

```python
names = []
```

```python
for i in thing:
    names.append(findName(i))
```
